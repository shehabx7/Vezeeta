<div class="alert alert-warning">
Vezeeta has 3 modes for stock status (out of stock, limited stocks, normal stock)

```python
vez_cleaning["StockStatus"] = vez_cleaning.stockLevelId.replace(1,"Out of stock").replace(2,"Limited stock").replace(3,"Normal")
vez_cleaning.drop(columns=["stockLevelId"],inplace=True)
vez_cleaning.reset_index(inplace=True,drop=True)
```



<div class="alert alert-warning">
Active ingerdints are contianed in a list of dictineories per product ,here i isolated it in a new data frame and normalized it.
</div>


```python
vez_cleaning["activeIngrediant"].replace("[]",0,inplace=True)
filterd_data = vez_cleaning[vez_cleaning["activeIngrediant"]  != 0]
filterd_data = filterd_data[["id","activeIngrediant"]]
filterd_data["activeIngrediant"] = filterd_data["activeIngrediant"].apply(literal_eval)
x = filterd_data.explode("activeIngrediant")
x.reset_index(drop=True,inplace=True)

filterd_data = pd.concat([x,pd.json_normalize(x["activeIngrediant"])],axis=1)
filterd_data
```



<div class="alert alert-warning">
Some products have the same active ingredients in arabic and in english, i filtered for english only as it is the main<br>
and produced normalized dataframe with each item id and it is ingredient name one by one.
</div>


```python
active_ing = filterd_data[filterd_data.lang != "ar"][["id","name"]]
active_ing 
```



<div class="alert alert-warning">


<div class="alert alert-warning">
  Repeating similar operation for the categories as we did for active ingredients, every product has a main category and list of categories that is is added to<br>, here i sepreated it as categoris per product
</div>


```python

sub = vez_cleaning.copy()
# sub.headCategoryKeys = sub.headCategoryKeys.tolist()
sub = sub[["id","headCategoryKeys"]]
sub.headCategoryKeys = sub.headCategoryKeys.apply(literal_eval)
x = pd.DataFrame(sub.headCategoryKeys.tolist(),columns=["main_cat","1","2","3"])
x.reset_index(drop=True,inplace=True)
x
sub = pd.concat([sub,x],axis=1)
sub.drop(columns="headCategoryKeys",inplace = True)
sub = pd.concat([sub,vez_cleaning["category"]],axis=1)
sub
```

```python
cat_key = sub[["main_cat","category"]].drop_duplicates("main_cat")

for column in sub.columns:
    for i in cat_key.main_cat.tolist():
       if i is not None:
            sub[column] = sub[column].replace(i,cat_key[cat_key.main_cat == i]["category"].tolist()[0])


sub.fillna("No",inplace=True)
sub.drop(columns=["main_cat"],inplace = True)
sub
```




<div class="alert alert-warning">
Our max for a product is to be added to 4 categories here i count product by number of categories is added to
</div>


```python

two_cat=sub[sub["1"] != "No"].groupby("1").id.count()
two_cat.index.set_names(["Category"],inplace=True)
two_cat = two_cat.reset_index()
two_cat.rename(columns={'id':'Count'},inplace=True)

three_cat=sub[sub["2"] != "No"].groupby("2").id.count()
three_cat.index.set_names(["Category"],inplace=True)
three_cat = three_cat.reset_index()
three_cat.rename(columns={'id':'Count'},inplace=True)


four_cat=sub[sub["3"] != "No"].groupby("3").id.count()
four_cat.index.set_names(["Category"],inplace=True)
four_cat = four_cat.reset_index()
four_cat.rename(columns={'id':'Count'},inplace=True)
```


<div class="alert alert-warning">
In the exploaation phase it was clear that the prices had outliers, here is did a box plot for all prices to see outliers,<br>
we can see the upper fence on the box plot and i also calculated it spereatly , <br>upper fence = IQR + 1.5 it is normal to consider values above it as outliers<br>
--then we did two histograms one for prices that we suspect to be outliers and one for the normal range
</div>


```python

import numpy as np


Q1 = np.percentile(vez_cleaning['newPrice'], 25, method='midpoint')
Q3 = np.percentile(vez_cleaning['newPrice'], 75, method='midpoint')
IQR = Q3 - Q1
upper = Q3+1.5*IQR



grouped = vez_cleaning[vez_cleaning.newPrice < upper][["id","newPrice"]]
grouped2 = vez_cleaning[vez_cleaning.newPrice > upper][["id","newPrice"]]


with_stock = vez_cleaning[vez_cleaning["StockStatus"] != "Out of stock"][["id","newPrice"]]

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

sns.histplot(grouped["newPrice"] ,ax=ax1)
ax1.set_title('Distropution of prices without outliers')
sns.histplot(grouped2["newPrice"] ,ax=ax2)
ax2.set_title('Distropution of prices of the outliers')


fig = px.box(with_stock, y="newPrice",title="Box plot of All prices")
fig.show()



```
